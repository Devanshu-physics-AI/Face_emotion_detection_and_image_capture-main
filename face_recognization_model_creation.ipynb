```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import time
import os
import tensorflow as tf
            
```

```python
#Getting file name in list for further processing
def load_images(which):
 generator=os.walk(f"/users/devanshusharma/desktop/images/train/{which}")
 file_path=[]
 for a,b,files in generator:
   initial=a+"/"
   for file in files:
    name=initial+file
    file_path.append(name)
 return file_path


    
    
 
    
```

```python
#load out image and convert it into the 200*200 dimension for more clearity of the image
def reshape(x):
    img=cv2.imread(x)
    img=cv2.resize(img,(224,224))
    return img
```

```python
#function for showing the image
def show_image(x):
    img=cv2.imread(str(x))
    plt.imshow(img)
    return 0

    
```

```python
#creating the csv file for path and the emotion as the output
#["anger","disgust","fear","happy","neutral","sad","surprise"]=[0,1,2,3,4,5,6]
angry=load_images("angry")
disgust=load_images("disgust")
fear=load_images("fear")
happy=load_images("happy")
neutral=load_images("neutral")
sad=load_images("sad")
surprise=load_images("surprise")
```

```python
#converting all into the data frames and adding their emotion values to
angry=pd.DataFrame(angry)
angry["emotion"]=np.nan
angry.fillna(0,inplace=True)
angry["emotion"]=angry["emotion"].astype(int)

```

```python
disgust=pd.DataFrame(disgust)
disgust["emotion"]=np.nan
disgust.fillna(1,inplace=True)
disgust["emotion"]=disgust["emotion"].astype(int)

```

```python
fear=pd.DataFrame(fear)
fear["emotion"]=np.nan
fear.fillna(2,inplace=True)
fear["emotion"]=fear["emotion"].astype(int)

```

```python
happy=pd.DataFrame(happy)
happy["emotion"]=np.nan
happy.fillna(3,inplace=True)
happy["emotion"]=happy["emotion"].astype(int)

```

```python
neutral=pd.DataFrame(neutral)
neutral["emotion"]=np.nan
neutral.fillna(4,inplace=True)
neutral["emotion"]=neutral["emotion"].astype(int)
neutral
```

```python
sad=pd.DataFrame(sad)
sad["emotion"]=np.nan
sad.fillna(5,inplace=True)
sad["emotion"]=sad["emotion"].astype(int)
```

```python
surprise=pd.DataFrame(surprise)
surprise["emotion"]=np.nan
surprise.fillna(6,inplace=True)
surprise["emotion"]=surprise["emotion"].astype(int)
```

```python
#concatinating all data and suffle the data
emotion_list=[angry,disgust,fear,happy,neutral,sad,surprise]
final_data=pd.concat(emotion_list)
```

```python
#checking the list that all the images is added or not
# sum=0
# for i in emotion_list:
#     sum=sum+len(i)
# print(sum)
```

```python
#suffling my final_data
np.random.seed(0)
final_data=final_data.sample(frac=0.6)
final_data=final_data.reset_index()
final_data
```

```python
#creating list of models
x=[] #list of all the images
y=[] #list of all the lebels
for i in range(0,len(final_data)):
    element=final_data["emotion"][i]
    y.append(element)
len(y)
```

```python
#Creating the list of numpy array of the data from the image path
def path_to_array(x):
    image=cv2.imread(str(x))
    return image
```

```python
#Adding the numpy array in the list

for path in final_data[0]:
  array=reshape(path)
  x.append(array)
x=np.array(x)
x.shape
```

```python
#functions I have made till now
#reshape     takes path as input and give the resized image
#show image   takes path as input and show image
```

```python
# plt.imshow(x[100])
# print(y[100])
```

```python
#normalise my model
x=x/255
```

```python
#training my model for the machine learning
model=tf.keras.applications.MobileNetV2()
```

```python
model.summary()
```

```python
#applying the transfer learning for my model
base_input=model.layers[1].input
base_output=model.layers[-2].output
```

```python
#making output as our requirements
from tensorflow.keras import layers
layer512=layers.Dense(512,activation="relu")(base_output)
# layer512=layers.Activation("relu")(layer512)
layer256=layers.Dense(256,activation="relu")(layer512)
# layer256=layers.Activation("relu")(layer256)
layer64=layers.Dense(64,activation="relu")(layer256)
# layer64=layers.Activation("relu")(layer64)
layer16=layers.Dense(16,activation="relu")(layer64)
layer16=layers.Activation("relu")(layer16)
final_output = layers.Dense(7, activation="softmax")(layer16)
```

```python
final_output
```

```python
from tensorflow import keras
final_model=keras.Model(inputs=base_input,outputs=final_output)
```

```python
final_model.summary()
```

```python
final_model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
x.shape
```

```python
print("model_train_complete")
final_model.fit(x,np.array(y),epochs=32)
# final_model.save("/users/devanshusharma/desktop/my_modela.h5")
```
